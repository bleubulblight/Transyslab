```python
# bs4의 BeautifulSoup를 가져온다.
from bs4 import BeautifulSoup
import requests
# selenium의 webdriver를 가져온다.
from selenium import webdriver
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.keys import Keys
from time import sleep
import collections
from openpyxl import Workbook

write_wb = Workbook()
```

```python
def opennewwindow(url) :
    driver.execute_script('window.open()')
    driver.switch_to.window(driver.window_handles[1])
    driver.get(url)
    
```

```python
def closewindow() :
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    sleep(0.5)
```

```python
# download 받은 chromedriver.exe 파일이 있는 path 입력 후, 중대본 webpage를 가져옴.
driver = webdriver.Chrome('C:/NeedDriver/chromedriver.exe')
driver.get('http://ncov.mohw.go.kr/tcmBoardList.do?brdId=3&brdGubun=')

# crawling을 위한 global variable 선언

pagenum = 1 # 페이지 번호 설정

while True :
    if pagenum % 10 != 0 :
            maintable = driver.find_element_by_xpath('//*[@id="content"]/div/div[4]/table/tbody')
            rows = maintable.find_elements_by_tag_name('tr')
            
            for index, value in enumerate(rows) :
                body = value.find_elements_by_tag_name('td')[1]
                a_tag = body.find_element_by_tag_name('a')
                _onclick = a_tag.get_attribute('onclick')
                _strings = _onclick.split(",")
                
                detailurl = ('http://ncov.mohw.go.kr/tcmBoardView.do?brdId={}&brdGubun={}&dataGubun=&ncvContSeq={}&contSeq=6543&board_id={}&gubun=ALL'
                             .format(_strings[1].strip("'"), _strings[2].strip("'"), _strings[3].strip("'"), _strings[4].strip("'")))
                
                opennewwindow(detailurl)
                
                _html = requests.get(detailurl).text
                soup = BeautifulSoup(_html)
                
                print(soup.select_one('p').get_text())
                print(soup.select_one('span').get_text())
                # 상세 페이지 내에서 크롤링
                
                closewindow()
                
            pageindex = (pagenum % 10) + 2
            wheretoclick = '//*[@id="content"]/div/div[5]/a[{}]'.format(pageindex)
            
            pagenum += 1
            
            clickpage = driver.find_element_by_xpath(wheretoclick)
            clickpage.click()
            sleep(1)
                
                
```
